##### Distribution and Univariate Analysis

### Student Info

```python
# print rows and columns
print(studentInfo.shape)

# Display the number of rows and columns, memory usage, column data types, and the count of non- null elements
print(studentInfo.info())

# Data Preview
print(studentInfo.head())

# Statistical Summary
print(studentInfo.describe(include='all'))

# check for missing values
print(studentInfo.isnull().sum())

# check for duplicate values
print(studentInfo.duplicated().sum())

# check for outliers
# check skewness. If skewness >+1 OR skewness <-1 then the data has outlier
print(studentInfo.skew())
x1=studentInfo.select_dtypes('number')
hist = x1.hist(bins=100)

# check for unique values and cardinality
for col in studentInfo:
    print(col, ":", studentInfo[col].nunique())
    print(col, ":", studentInfo[col].unique())
```
