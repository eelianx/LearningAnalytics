```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from matplotlib import pyplot
# extract only the numeric ones + the target variable
dataset_pre=EEE_Week16[['num_of_prev_attempts','studied_credits','sum_weighted_score',\
                            'sum_score','count_total_assessment','count_passed_assessment_TMA',\
                            'count_passed_assessment_Total','count_failed_assessment_TMA',\
                            'count_failed_assessment_Total','rate_passed_assessment',\
                            'sum_click_total','sum_click_total_precoursestart',\
                            'sum_click_resource','sum_click_oucontent','sum_click_url',\
                            'sum_click_homepage','sum_click_subpage','sum_click_forumng',\
                            'sum_click_oucollaborate','sum_click_quiz','sum_click_page',\
                            'sum_click_ouwiki','sum_click_dualpane',\
                            'final_weighted_score','_flag_final_result']]
# retrieve numpy array
dataset = dataset_pre.values
# split into input (X) and output (y) variables
X = dataset[:, :-1]
y = dataset[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# feature selection
X_train_fs, X_test_fs, fs = select_features(X_train, y_train, X_test)
# what are scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
pyplot.bar([i for i in range(len(fs.scores_))], fs.scores_)
pyplot.show()
```
