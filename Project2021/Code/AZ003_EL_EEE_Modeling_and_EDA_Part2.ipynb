* pvalue>0.05 = accept null hypothesis and feature is not important
* pvalue<0.05 = reject null hypothesis and feature is important

```python
# Chi square Test

categories =['gender', 'region', 'highest_education', 'imd_band', 'age_band', 'disability', 'final_result']

y=EEE_Week16['_flag_final_result']
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2
for col in categories:
    xcat = LabelEncoder().fit_transform(EEE_Week16[col]).reshape(-1,1)
    cval,pval=chi2(xcat,y)
    print(col,pval)
```

* pvalue>0.05 = accept null hypothesis and feature is not important
* pvalue<0.05 = reject null hypothesis and feature is important

### Logistic Regression

```python
# data prep
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder,StandardScaler
preprocessor = ColumnTransformer([('ohe',OneHotEncoder(),[0,1]),\
                                  #('ode',OrdinalEncoder(),[0]),\
                                  ('sc',StandardScaler(),slice(2,15))\
                                 ],remainder="passthrough")
xtotal_m1 = preprocessor.fit_transform(xtotal_prepcs)
ytotal_m1 = ytotal_prepcs
#pd.DataFrame(xtotal_m1).head()
xtotal_xm1=np.delete(xtotal_m1, -1, axis=1)
ytotal_xm1=ytotal_prepcs['_flag_final_result']
xtrain_m1 = xtotal_m1[xtotal_m1[:, -1] == 'TRAIN']
xtrain_m1 = np.delete(xtrain_m1, -1, axis=1)
xtest_m1 = xtotal_m1[xtotal_m1[:, -1] == 'TEST']
xtest_m1 = np.delete(xtest_m1, -1, axis=1)

# check sizes
print(xtotal_prepcs.shape)
print(xtotal_m1.shape)
print(xtrain_m1.shape)
print(xtest_m1.shape)
print(ytotal_prepcs.shape)
print(ytotal_m1.shape)
print(ytrain_m1.shape)
print(ytest_m1.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder,StandardScaler
preprocessor = ColumnTransformer([('ohe',OneHotEncoder(),[0,1])],\
                                  #('ode',OrdinalEncoder(),[0]),\
                                  remainder="passthrough")

ytrain_mm2= ytotal_prepcs[(ytotal_prepcs['_filter'].isin(['TRAIN']))]
ytest_mm2= ytotal_prepcs[(ytotal_prepcs['_filter'].isin(['TEST']))]
ytrain_m2=ytrain_mm2['_flag_final_result']
ytest_m2=ytest_mm2['_flag_final_result']

xtotal_m2 = preprocessor.fit_transform(xtotal_prepcs)
ytotal_m2 = ytotal_prepcs
xtrain_m2 = xtotal_m2[xtotal_m2[:, -1] == 'TRAIN']
xtrain_m2 = np.delete(xtrain_m2, -1, axis=1)
xtest_m2 = xtotal_m2[xtotal_m2[:, -1] == 'TEST']
xtest_m2 = np.delete(xtest_m2, -1, axis=1)

xtotal_xm2=np.delete(xtotal_m2, -1, axis=1)
ytotal_xm2=ytotal_prepcs['_flag_final_result']

print(xtrain_m2.shape)
print(xtest_m2.shape)
print(ytrain_m2.shape)
print(ytest_m2.shape)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model_m2=DecisionTreeClassifier(random_state=5)
model_m2.fit(xtrain_m2,ytrain_m2)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score
ypredtrain_m2 = model_m2.predict(xtrain_m2)
print("model_m2_train")
print("Accuracy: ",metrics.accuracy_score(ytrain_m2,ypredtrain_m2))
print("Recall: ",metrics.recall_score(ytrain_m2,ypredtrain_m2))
print("Precision: ",metrics.precision_score(ytrain_m2,ypredtrain_m2))
print("F1 Score: ",metrics.f1_score(ytrain_m2,ypredtrain_m2))
print("AUC: ", roc_auc_score(ytrain_m2, ypredtrain_m2))
print(metrics.confusion_matrix(ytrain_m2,ypredtrain_m2))
print("model_m2_test")
print("")
#from sklearn import metrics
ypredtest_m2 = model_m2.predict(xtest_m2)
print("Accuracy: ",metrics.accuracy_score(ytest_m2,ypredtest_m2))
print("Recall: ",metrics.recall_score(ytest_m2,ypredtest_m2))
print("Precision: ",metrics.precision_score(ytest_m2,ypredtest_m2))
print("F1 Score: ",metrics.f1_score(ytest_m2,ypredtest_m2))
print("AUC: ", roc_auc_score(ytest_m2, ypredtest_m2))
print(metrics.confusion_matrix(ytest_m2,ypredtest_m2))
```
